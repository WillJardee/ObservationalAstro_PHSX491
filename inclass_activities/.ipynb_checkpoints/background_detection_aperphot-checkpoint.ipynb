## Create a synthetic image

```python
from astropy.modeling.models import Gaussian2D
from photutils.datasets import make_noise_image
```

```python
def fake_image(fwhm,nstars=10,npix=200):
    """ x0, y0, amplitude, asemi, bsemi, theta (deg), npix wide"""
    data = np.zeros((npix,npix),float)
    y, x = np.mgrid[0:npix, 0:npix]
    ny = nx = npix
    for i in range(nstars):
        x0,y0 = np.random.rand(2)*(npix-1)
        amplitude = 10**(np.random.rand(1)*3.7)+50
        g = Gaussian2D(amplitude, x0, y0, fwhm/2.35, fwhm/2.35, theta=0)
        data += g(x, y)
    # Add sky
    data += 0.3*x + 0.002*x*y - 0.5*y + 555
    # Add noise
    data += np.random.randn(npix,npix)*np.sqrt(data)
    return data
```

```python
data = fake_image(5,50)
```

```python
fig = plt.figure()
plt.imshow(data,origin='lower',vmin=500,vmax=800)
plt.colorbar()
plt.show()
```

# Simple Source Detection

Now we just need an automated way of detecting our stars in the image.

The pixels with stars in them have higher flux than the pixels with only background.  Let's set a threshold level and find all the stars that have flux above this level.

```python
# Determine robust mean, median and standard deviation of the image
mean, median, std = sigma_clipped_stats(sim, sigma=3.0)  
print((median, std))  
```

```python
# We'll set a threshold of 5*sigma above the background
thresh = median + 5*std
print('Thresh = '+str(thresh))
```

```python
# Get a boolean array (True/False) of which pixels are above/below the threshold
det = (sim > thresh)
```

```python
# This is what it looks like
fit = plt.figure(figsize=(15,15))
plt.imshow(np.array(det).astype(float),origin='lower')
plt.colorbar()
plt.xlim(0,600)
plt.ylim(0,600)
```

This is picking out the "star" pixels, but we are getting lots of pixels per star.  We really just want one pixel per star so we can create a list or table of those.

The central pixel of a star should be a "peak", i.e. it's flux is higher than those of the surrounding pixels, a local maximum.  One easy way to calculate this is to compare a pixel to it's neighbors by shifting the image by one pixel left, right, up and down.  We can use np.roll() to do this.  Then we require that the pixel is brighter than it's neighbors.

```python
sim1 = np.roll(sim,1,axis=0)    # shift right 1 pixel
sim2 = np.roll(sim,-1,axis=0)   # shift left 1 pixel
sim3 = np.roll(sim,1,axis=1)    # shift up 1 pixel
sim4 = np.roll(sim,-1,axis=1)   # shift down 1 pixel
# Only want pixels that are above the flux threshold and is a local maximum (brighter than it's neighbors)
det2 = (sim > thresh) & (sim>sim1) & (sim>sim2) & (sim>sim3) & (sim>sim4)
```

```python
# Let's plot that up
fig = plt.figure(figsize=(15,15))
plt.imshow(np.array(det2).astype(float),origin='lower')
plt.colorbar()
plt.xlim(0,600)
plt.ylim(0,600)
```

This has selected less pixels per star, but still too many.  That's because there is noise in the image and this creates lots of local maxima.

One way to get around this is to smooth our image so that the central pixel really is the local maximum.  A rule of thumb is to use a smoothing size of roughly our seeing, which in our case is ~10 pixels.

```python
# We will smooth with a 2D Gaussian
# use fwhm=10 which converts to sigma=fwhm/2.35 = 10/2.35 = 4.3
smsim = ndimage.gaussian_filter(sim, sigma=(4.3, 4.3), order=0)
```

```python
# We can see that the image now looks much smoother
fig = plt.figure(figsize=(15,15))
plt.imshow(smsim,origin='lower',vmin=0,vmax=1000)
plt.colorbar()
plt.xlim(0,600)
plt.ylim(0,600)
```

```python
# The smoothing also reduces the noise in the image, so we need to modify our flux threshold accordingly
mean2, median2, std2 = sigma_clipped_stats(smsim, sigma=3.0)  
print((median2, std2))  
# The median didn't change much, but the std.dev. went from 59 to 11
```

```python
# And our new treshold
thresh2 = median2 + 5*std2
print('Thresh = '+str(thresh2))
```

```python
# Let's update our code using the smoothed image
smsim1 = np.roll(smsim,1,axis=0)
smsim2 = np.roll(smsim,-1,axis=0)
smsim3 = np.roll(smsim,1,axis=1)
smsim4 = np.roll(smsim,-1,axis=1)
det3 = (smsim > thresh2) & (smsim>smsim1) & (smsim>smsim2) & (smsim>smsim3) & (smsim>smsim4)
```

```python
# Let's visualize the new detection image.  It's hard to see, but there are just a handful of pixels selected
fig = plt.figure(figsize=(15,15))
plt.imshow(np.array(det3).astype(float),origin='lower')
plt.colorbar()
plt.xlim(0,600)
plt.ylim(0,600)
```

This is now doing exactly what we want!  It is selecting only the central pixel for each star

```python
# Use np.where() to get the positions of the pixels that we selected
yind, xind = np.where(det3 == True)
```

```python
# This is how many stars we detected
print(len(yind))
```

```python
# Let's plot up our detections
fig = plt.figure(figsize=(15,15))
plt.imshow(sim,vmin=-100,vmax=1000)
plt.xlim(0,2000)
plt.ylim(0,2000)
plt.scatter(xind,yind,c='r',marker='+')
plt.colorbar()
```

```python
# Let's wrap it all up in a function that takes the smoothing length (fwhm) and the number of
# standard deviations of the background scatter in the threshold (nsig)
def detection(im,fwhm=5,nsig=5):
    # Smooth the image
    smim = ndimage.gaussian_filter(im, sigma=(fwhm/2.35, fwhm/2.35), order=0)
    # Calculate the median and scatter
    mean, median, std = sigma_clipped_stats(smim, sigma=3.0)
    # Shift the images 4 ways
    smim1 = np.roll(smim,1,axis=0)
    smim2 = np.roll(smim,-1,axis=0)
    smim3 = np.roll(smim,1,axis=1)
    smim4 = np.roll(smim,-1,axis=1)
    # Set the threshold
    thresh = median + nsig*std
    # Do the detection
    det = (smim > thresh) & (smim>smim1) & (smim>smim2) & (smim>smim3) & (smim>smim4)
    ind1, ind2 = np.where(det == True)
    # Make a table
    dtype = np.dtype([('xpos',float),('ypos',float)])
    cat = np.zeros(len(ind1),dtype=dtype)
    cat['xpos'] = ind2
    cat['ypos'] = ind1
    return cat
```

While our algorithm did a great job, there are still some faint stars visible in the image
that we didn't pick up.  Let's try to lower our detection threshold to see if we can pick them up.

```python
# Lower to 3 sigma
cat = detection(sim,10,3)
```

```python
# That's a lot more
len(cat)
```

```python
# Let's plot our image and the detected sources
fig = plt.figure(figsize=(15,15))
plt.imshow(sim,vmin=-100,vmax=500)
plt.xlim(0,2000)
plt.ylim(0,2000)
plt.scatter(cat['xpos'],cat['ypos'],c='r',marker='+')
plt.colorbar()
```

```python
# At some point we will start to get some spurious detections
# from noise spikes that look like stars
cat2 = detection(sim,10,1.5)
```

```python
len(cat2)
```

```python
fig = plt.figure(figsize=(15,15))
plt.imshow(sim,vmin=-100,vmax=500)
plt.xlim(0,2000)
plt.ylim(0,2000)
plt.scatter(cat2['xpos'],cat2['ypos'],c='r',marker='+')
plt.colorbar()
```

That's a lot of spurious sources. You need to evaluate what threshold works best for you.

A lower threshold will have more of the stars (more complete), but more spurious sources (higher contamination).
A higher threshold will have less stars (less complete), but also less spurious sources (lower contamination and higher fidelity).

## Let's make a function to do these steps

```python
# I created a function that will do the steps for us
#  just input the image (im), positions, aperture radius (r) and sky inner (rin) and outer (rout) radius.
def aper_phot(im,positions,r=None,rin=None,rout=None):
    # Define the aperture right around our star
    aperture = CircularAperture(positions, r=r)
    # Define the sky background circular annulus aperture
    annulus_aperture = CircularAnnulus(positions, r_in=rin, r_out=rout)
    # This turns our sky background aperture into a pixel mask that we can use to calculate the median value
    annulus_masks = annulus_aperture.to_mask(method='center')
    # Measure the median background value for each star
    bkg_median = []
    for mask in annulus_masks:  # loop over the stars
        # Get the data in the annulus
        annulus_data = mask.multiply(im)
        annulus_data_1d = annulus_data[mask.data > 0]                # Only want positive values
        _, median_sigclip, _ = sigma_clipped_stats(annulus_data_1d)  # calculate median
        bkg_median.append(median_sigclip)                            # add to our median list
    bkg_median = np.array(bkg_median)                                # turn into numpy array
    # Calculate the aperture photometry
    phot = aperture_photometry(im, aperture)
    # Stuff it in a table
    phot['annulus_median'] = bkg_median
    phot['aper_bkg'] = bkg_median * aperture.area                      # the number of bkg counts in our aperture
    phot['aper_flux'] = phot['aperture_sum'] - phot['aper_bkg']  # subtract bkg contribution
    phot['mag'] = -2.5*np.log10(phot['aper_flux'].data)+25
    return phot
```

```python
# Now run it on our positions
phot = aper_phot(sim,positions,r=20,rin=35,rout=55)
```
